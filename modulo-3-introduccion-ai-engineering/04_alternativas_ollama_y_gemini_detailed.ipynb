## 1. ¿Por qué considerar alternativas a OpenAI?

OpenAI es el _default_ pero rara vez es la única opción razonable. Razones para mirar alternativas:

| Razón | Explicación |
|---|---|
| **Privacidad / regulación** | Tus datos (médicos, legales, financieros) no pueden salir de tu infraestructura. Alternativas serias incluyen self-hosting con vLLM/TGI o cloud regional. |
| **Costo** | Volumen alto + tarea simple → un modelo open-source vía Groq/Cerebras o un Haiku/Flash sale 10-100× más barato. |
| **Latencia** | Inferencia en hardware especializado (Groq LPU, Cerebras WSE) = 5-10× más rápida que GPU típica. |
| **Vendor lock-in** | Diversificar proveedores te protege de cambios de pricing, deprecaciones de modelos, downtime. |
| **Customización** | Open-source te deja fine-tunear, cuantizar, modificar la arquitectura. |
| **Frontera** | A veces _no_ es OpenAI quien tiene el mejor modelo para tu tarea (Claude para código, Gemini para multimodal de contexto largo). |

En este notebook cubrimos cuatro alternativas representativas:
- **Gemini** — la familia de modelos de Google, con context window enorme (1-2M tokens) y multimodal nativo.
- **Groq** — inferencia ultra-rápida de modelos open-source sobre LPUs.
- **Cerebras** — inferencia ultra-rápida sobre Wafer-Scale Engine, throughput líder en modelos grandes.
- **Ollama Cloud** — modelos open-source MUY grandes (DeepSeek 671B, Qwen3 480B, Kimi K2 1T) sin necesidad de GPU local.

## 4. Ollama Cloud — modelos open-source grandes desde la nube

[Ollama](https://ollama.com/) es conocido por ser el "Docker para LLMs" — un runtime para correr Llama, Mistral, Qwen, Gemma, etc. Localmente funciona genial, pero correr modelos grandes (>70B) requiere hardware caro.

**Ollama Cloud** resuelve eso: la **misma CLI y misma librería** que usarías en local funcionan contra modelos servidos desde la infraestructura de Ollama, incluyendo modelos demasiado grandes para tu Mac/PC (DeepSeek 671B, Qwen3 480B, GPT-OSS 120B, Kimi K2 1T).

**Características clave:**
- API HTTP **compatible con OpenAI** — puedes apuntar el SDK de OpenAI a Ollama cambiando solo `base_url`.
- **Misma interfaz** que Ollama local: si mañana quisieras moverte a self-hosting, el código no cambia.
- Tier gratuito + planes pagos. Sin tarjeta para empezar.
- **No necesitas GPU local** — ideal para talleres, prototipado, demos.

> 💡 En este curso usamos **solo el modo cloud** para simplificar el setup. Si te interesa correr modelos en tu máquina, instala [Ollama](https://ollama.com/download) localmente: la API es idéntica, solo cambia el `host` del cliente.

### 4.1 Setup

```bash
# 1. Crear key gratis en https://ollama.com/settings/keys
# 2. Agregar a ai.env:  OLLAMA_API_KEY=...
# 3. Instalar la librería oficial
uv add ollama
```

La librería de Python es la misma para local y cloud — solo cambia el `host` y los headers de Authorization.

````python
# Chat con Ollama Cloud — la librería ollama apuntada a https://ollama.com
# uv add ollama
import os
from ollama import Client

ollama_cloud = Client(
    host='https://ollama.com',
    headers={'Authorization': f"Bearer {os.environ.get('OLLAMA_API_KEY')}"},
)

resp = ollama_cloud.chat(
    model='gpt-oss:120b-cloud',     # sufijo -cloud → Ollama Cloud
    messages=[
        {'role': 'system', 'content': 'Eres un asistente conciso. Responde en máximo 3 frases.'},
        {'role': 'user',   'content': '¿Qué es un Transformer? pero los de las peliculas'},
    ],
    options={
        'temperature': 0.3,
        'num_predict': 200,        # equivalente a max_tokens
    },
)
print(resp['message']['content'])
print('---')
print(f"Tokens evaluados: {resp.get('eval_count', 'n/a')}")
````

````python
# Streaming token a token con Ollama Cloud
stream = ollama_cloud.chat(
    model='qwen3-coder:480b-cloud',
    messages=[{'role': 'user', 'content': 'Escribe una función Python que invierta una lista enlazada.'}],
    stream=True,
)

for chunk in stream:
    print(chunk['message']['content'], end='', flush=True)
````

## 6. Groq — inferencia ultra-rápida con LPUs

[Groq](https://groq.com/) sirve modelos open-source corriendo sobre hardware **LPU (Language Processing Unit)** — chips diseñados específicamente para inferencia de LLMs. Resultado: latencias y throughput **muy superiores** a GPUs convencionales.

| | Groq (LPU) | GPU típica en cloud |
|---|---|---|
| Throughput | **300-800 tokens/s** | 30-100 tokens/s |
| Latencia primer token | ~50 ms | 200-500 ms |
| Modelos | Llama 3.1/3.3, Mixtral, Gemma, Qwen, DeepSeek R1 | depende del proveedor |

**Por qué importa:** en aplicaciones interactivas (chat, agentes, voz), la velocidad cambia la sensación del producto. Una respuesta que tarda 500 ms vs 5 s es la diferencia entre "instantáneo" y "esperando".

**Otras ventajas:**
- **Tier gratuito generoso** — 30 RPM en la mayoría de modelos, sin tarjeta.
- **API compatible con OpenAI** — solo cambias `base_url` y `api_key`. Mismo truco que con Ollama.
- **Modelos open-source** — no estás casado con un modelo propietario; si Groq mañana sube precios, te llevas el código y lo corres en otro proveedor con los mismos pesos.

### 6.1 Setup

```bash
# 1. Crear key gratis en https://console.groq.com/keys
# 2. Agregar a ai.env:  GROQ_API_KEY=gsk_...
# 3. El SDK de OpenAI ya está instalado, no hace falta uno nuevo.
```

La key ya quedó cargada en la sección 2 desde `ai.env`.

````python
# Groq usando el SDK de OpenAI — el truco está en base_url
import os, time
from openai import OpenAI

groq = OpenAI(
    base_url='https://api.groq.com/openai/v1',
    api_key=os.environ.get('GROQ_API_KEY'),
)

t0 = time.time()
resp = groq.chat.completions.create(
    model='llama-3.3-70b-versatile',
    messages=[
        {'role': 'system', 'content': 'Eres un asistente conciso. Responde en español, máximo 3 frases.'},
        {'role': 'user',   'content': 'Resume qué es Groq y por qué importa.'},
    ],
    temperature=0.3,
    max_tokens=150,
)
elapsed = time.time() - t0

print(resp.choices[0].message.content)
print('---')
print(f'Tokens output: {resp.usage.completion_tokens}')
print(f'Tiempo total : {elapsed:.2f}s  →  {resp.usage.completion_tokens/elapsed:.0f} tok/s')
````

### 6.2 Streaming con Groq

Aquí es donde se nota la velocidad — el texto aparece tan rápido que apenas alcanzas a leerlo. Útil para experiencias tipo "instantáneo".

````python
# Streaming con Groq — fíjate en la velocidad de tokens/segundo
import time

t0 = time.time()
first_token_at = None
total_tokens = 0

stream = groq.chat.completions.create(
    model='llama-3.3-70b-versatile',
    messages=[{'role': 'user', 'content': 'Cuenta del 1 al 10 con una palabra creativa por número.'}],
    stream=True,
    stream_options={'include_usage': True},   # último chunk trae usage real
)

for chunk in stream:
    if chunk.usage:
        total_tokens = chunk.usage.completion_tokens
    if not chunk.choices:
        continue
    delta = chunk.choices[0].delta.content
    if delta:
        if first_token_at is None:
            first_token_at = time.time() - t0
        print(delta, end='', flush=True)

elapsed = time.time() - t0
print(f'\n\n---')
print(f'Time to first token : {first_token_at*1000:.0f} ms')
print(f'Tiempo total        : {elapsed:.2f} s')
print(f'Tokens generados    : {total_tokens}')
print(f'Throughput          : {total_tokens/elapsed:.0f} tok/s')
````

## 6.4 Arquitecturas de hardware para inferencia — GPU vs LPU vs WSE

### El problema con GPUs para inferencia

Las GPUs (NVIDIA A100, H100) fueron diseñadas para **entrenamiento** — operaciones masivas de matriz-matriz con alta utilización de cómputo. Pero la **inferencia** de LLMs tiene un perfil diferente:

| Fase | Operación | Bottleneck |
|---|---|---|
| **Prefill** (procesar input) | Matriz-matriz (paralelo) | Cómputo (FLOPS) |
| **Decode** (generar output) | Matriz-vector (secuencial) | **Memoria (bandwidth)** |

En decode, la GPU está **ociosa en cómputo** la mayor parte del tiempo, esperando que los datos lleguen desde HBM (High Bandwidth Memory). Esto se llama **memory-bound inference**.

**Métricas clave:**
- **FLOPS (operaciones/segundo):** GPU gana
- **Bandwidth (GB/s):** Aquí es donde LPU y WSE brillan
- **Latencia de memoria:** Crítica para time-to-first-token

### Groq LPU (Language Processing Unit)

**Arquitectura:**
- Chip ASIC diseñado específicamente para inferencia de Transformers
- **Memoria on-chip masiva:** ~230 MB de SRAM (vs ~40 MB en A100)
- **Deterministic execution:** No hay cache misses — todo es predecible
- **Dataflow architecture:** Los datos fluyen directamente entre unidades sin pasar por memoria externa

**Ventajas:**
- **Bandwidth extremo:** ~80 TB/s interno (vs ~2 TB/s en A100 HBM)
- **Latencia predecible:** No hay variabilidad por cache misses
- **Time-to-first-token bajo:** ~50 ms (vs 200-500 ms en GPU)

**Limitaciones:**
- Solo modelos que caben en los 230 MB de SRAM (hasta ~70B parámetros con cuantización)
- No es programable como GPU — solo para inferencia de Transformers

### Cerebras WSE (Wafer-Scale Engine)

**Arquitectura:**
- **Un chip del tamaño de una oblea entera:** 46,225 mm² (vs ~826 mm² en H100)
- **850,000 cores** en un solo chip
- **40 GB de SRAM on-chip** (vs ~40 MB en A100)
- **20 PB/s de bandwidth interno**

**Ventajas:**
- **Throughput absurdo:** >2,000 tok/s en Llama 3.1 70B
- **Escalabilidad:** Modelos de hasta 1T parámetros en un solo chip
- **Sin comunicación inter-chip:** Todo está en la misma oblea

**Limitaciones:**
- **Costo:** Un WSE cuesta millones de dólares
- **Consumo energético:** ~20 kW por chip
- **Yield:** Fabricar obleas sin defectos es difícil

### Comparación de arquitecturas

| Métrica | GPU (H100) | Groq LPU | Cerebras WSE |
|---|---|---|---|
| **Área del chip** | 826 mm² | ~750 mm² | 46,225 mm² |
| **SRAM on-chip** | 50 MB | 230 MB | 40 GB |
| **Bandwidth interno** | 3.35 TB/s | 80 TB/s | 20 PB/s |
| **Cores** | 16,896 CUDA | ~1,000 tiles | 850,000 |
| **FLOPS (FP16)** | 989 TFLOPS | ~750 TFLOPS | ~1 EFLOPS |
| **Consumo** | 700W | ~300W | ~20 kW |
| **Precio** | ~$30K | N/A (cloud) | ~$2-3M |
| **Throughput (Llama 70B)** | ~30-100 tok/s | ~280 tok/s | ~2,200 tok/s |
| **Time-to-first-token** | 200-500 ms | ~50 ms | ~100 ms |

### ¿Por qué no todos usan LPU/WSE?

**Costo de desarrollo:**
- Diseñar un ASIC cuesta $100M-500M
- Solo vale la pena si vendes millones de chips o servicios a escala masiva

**Flexibilidad:**
- GPU puede entrenar, hacer inferencia, renderizar gráficos, minar crypto
- LPU/WSE solo hacen inferencia de Transformers

**Ecosistema:**
- CUDA (NVIDIA) tiene 15 años de herramientas, librerías, optimizaciones
- LPU/WSE requieren compiladores y runtimes propios

**Disponibilidad:**
- GPUs se pueden comprar/rentar en cualquier cloud
- LPU/WSE solo vía Groq Cloud y Cerebras Cloud

### El futuro: hardware especializado

La tendencia es clara: **inferencia de LLMs se moverá a ASICs especializados**. Razones:

1. **Eficiencia energética:** LPU/WSE hacen 5-10x más trabajo por watt
2. **Costo a escala:** A millones de chips, el ASIC es más barato que GPU
3. **Latencia:** Crítica para aplicaciones interactivas

**Otros jugadores:**
- **Google TPU v5/v6:** Diseñados para entrenamiento e inferencia de Transformers
- **AWS Trainium/Inferentia:** ASICs de Amazon para ML
- **Microsoft Maia:** Chip propio para Azure
- **Startups:** Etched (Sohu), Tenstorrent, SambaNova

En 5 años, es probable que la mayoría de la inferencia de producción corra en ASICs, no en GPUs.

````python
# Comparación visual de arquitecturas de hardware
import matplotlib.pyplot as plt
import numpy as np

# Datos de comparación
arquitecturas = ['GPU\\n(H100)', 'Groq\\nLPU', 'Cerebras\\nWSE']

# Métricas normalizadas (GPU = 1.0 como baseline)
metricas = {
    'SRAM on-chip (MB)': [50, 230, 40000],
    'Bandwidth (TB/s)': [3.35, 80, 20000],
    'Throughput\\n(tok/s Llama 70B)': [65, 280, 2200],
    'Time-to-first-token (ms)': [350, 50, 100],  # menor es mejor
}

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

colores = ['steelblue', 'orange', 'green']

for idx, (metrica, valores) in enumerate(metricas.items()):
    ax = axes[idx]
    
    # Para time-to-first-token, invertimos (menor es mejor)
    if 'time' in metrica.lower():
        # Mostrar como "velocidad" (inverso)
        valores_plot = [1000/v for v in valores]
        ylabel = 'Velocidad (1/ms)'
        titulo = metrica.replace('Time-to-first-token (ms)', 'Velocidad de respuesta\\n(mayor es mejor)')
    else:
        valores_plot = valores
        ylabel = 'Valor'
        titulo = metrica
    
    bars = ax.bar(arquitecturas, valores_plot, color=colores, alpha=0.8, edgecolor='white')
    ax.set_ylabel(ylabel)
    ax.set_title(titulo, fontsize=11, fontweight='bold')
    ax.grid(True, alpha=0.3, axis='y')
    
    # Escala logarítmica para métricas con gran rango
    if max(valores_plot) / min(valores_plot) > 100:
        ax.set_yscale('log')
    
    # Anotar valores originales
    for bar, val_original in zip(bars, valores):
        height = bar.get_height()
        if 'time' in metrica.lower():
            label = f'{val_original:.0f}ms'
        elif 'Bandwidth' in metrica:
            if val_original >= 1000:
                label = f'{val_original/1000:.0f}PB/s'
            else:
                label = f'{val_original:.1f}TB/s'
        elif 'SRAM' in metrica:
            if val_original >= 1000:
                label = f'{val_original/1000:.0f}GB'
            else:
                label = f'{val_original:.0f}MB'
        else:
            label = f'{val_original:.0f}'
        
        ax.text(bar.get_x() + bar.get_width()/2., height * 1.1,
                label, ha='center', va='bottom', fontsize=9, fontweight='bold')

plt.suptitle('Comparación de arquitecturas de hardware para inferencia de LLMs', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Tabla comparativa de eficiencia
print("="*70)
print("EFICIENCIA RELATIVA (GPU H100 = 1.0x baseline)")
print("="*70)

baseline_throughput = 65  # tok/s en GPU
baseline_ttft = 350  # ms en GPU

for i, arch in enumerate(arquitecturas):
    throughput = metricas['Throughput\\n(tok/s Llama 70B)'][i]
    ttft = metricas['Time-to-first-token (ms)'][i]
    
    speedup_throughput = throughput / baseline_throughput
    speedup_ttft = baseline_ttft / ttft  # inverso porque menor es mejor
    
    print(f"\n{arch.replace(chr(10), ' ')}:")
    print(f"  Throughput:  {speedup_throughput:>5.1f}x más rápido")
    print(f"  Latencia:    {speedup_ttft:>5.1f}x más rápido")
    print(f"  SRAM:        {metricas['SRAM on-chip (MB)'][i] / 50:>5.0f}x más memoria on-chip")

# Visualización de área del chip
fig, ax = plt.subplots(figsize=(10, 6))

areas = [826, 750, 46225]  # mm²
areas_normalized = [np.sqrt(a) for a in areas]  # para visualizar como cuadrados

positions = [0, 1.5, 5]
for i, (arch, area, pos) in enumerate(zip(arquitecturas, areas, positions)):
    size = areas_normalized[i] / 100  # escalar para visualización
    rect = plt.Rectangle((pos, 0), size, size, 
                         facecolor=colores[i], alpha=0.7, edgecolor='black', linewidth=2)
    ax.add_patch(rect)
    
    # Etiqueta
    ax.text(pos + size/2, -0.3, arch.replace('\n', ' '), 
            ha='center', fontsize=11, fontweight='bold')
    ax.text(pos + size/2, size + 0.2, f'{area:,} mm²', 
            ha='center', fontsize=10)

ax.set_xlim(-0.5, 8)
ax.set_ylim(-0.8, 3)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title('Tamaño relativo de los chips (área en mm²)', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()

print("\n💡 Observaciones clave:")
print("   • WSE tiene 56x más área que H100 (una oblea entera vs un chip)")
print("   • LPU tiene 4.6x más SRAM on-chip que GPU → menos accesos a memoria externa")
print("   • Throughput: WSE 34x > GPU, LPU 4.3x > GPU")
print("   • Latencia: LPU 7x < GPU, WSE 3.5x < GPU")
print("   • Trade-off: Especialización vs Flexibilidad")
````

## 10. Patrón de abstracción del proveedor

En producción, **no acoples tu lógica a un SDK concreto**. Encapsula las llamadas detrás de una pequeña interfaz para poder cambiar de proveedor sin tocar el resto del código:

```python
class LLMClient:
    def chat(self, messages, **opts) -> str: ...

class OpenAIClient(LLMClient):
    def __init__(self):
        self.client = OpenAI()  # usa OPENAI_API_KEY del entorno
    def chat(self, messages, **opts):
        return self.client.chat.completions.create(
            model='gpt-4o-mini', messages=messages, **opts
        ).choices[0].message.content

class GroqClient(LLMClient):
    def __init__(self):
        self.client = OpenAI(base_url='https://api.groq.com/openai/v1',
                             api_key=os.environ['GROQ_API_KEY'])
    def chat(self, messages, **opts):
        return self.client.chat.completions.create(
            model='llama-3.3-70b-versatile', messages=messages, **opts
        ).choices[0].message.content

class CerebrasClient(LLMClient):
    def __init__(self):
        self.client = OpenAI(base_url='https://api.cerebras.ai/v1',
                             api_key=os.environ['CEREBRAS_API_KEY'])
    def chat(self, messages, **opts):
        return self.client.chat.completions.create(
            model='llama-3.3-70b', messages=messages, **opts
        ).choices[0].message.content

class OllamaCloudClient(LLMClient):
    def __init__(self):
        self.client = OpenAI(base_url='https://ollama.com/v1',
                             api_key=os.environ['OLLAMA_API_KEY'])
    def chat(self, messages, **opts):
        return self.client.chat.completions.create(
            model='gpt-oss:120b-cloud', messages=messages, **opts
        ).choices[0].message.content

class GeminiClient(LLMClient): ...  # usa google-genai (no es compatible OpenAI)
```

Beneficios:
- **A/B testing** de proveedores por feature flag.
- **Fallback** automático si un proveedor cae (ej. si OpenAI 429, te vas a Groq → Cerebras).
- **Tests** con un fake `LLMClient` que devuelve respuestas determinísticas.
- **Migración** entre proveedores sin reescribir lógica de negocio.

> 🎯 Nota: como **Groq, Cerebras y Ollama Cloud** ya exponen API compatible OpenAI, sus `chat()` son básicamente idénticos al de OpenAI — la abstracción te sale casi gratis.

## 9.1 Estrategias avanzadas de selección de proveedor

### Framework de decisión: 5 dimensiones

**1. Calidad del modelo**

No todos los modelos son iguales para todas las tareas:

| Tarea | Mejor modelo (2026) | Proveedor |
|---|---|---|
| Código complejo | Claude Opus 4, GPT-5 | Anthropic, OpenAI |
| Razonamiento matemático | o3, DeepSeek R1, Claude Opus 4 | OpenAI, open (Groq/Cerebras), Anthropic |
| Escritura creativa | Claude Opus 4, GPT-5 | Anthropic, OpenAI |
| Multimodal (video) | Gemini 2.5 Pro | Google |
| Contexto ultra-largo | Gemini 2.5 Pro (2M), Claude Opus 4 (1M) | Google, Anthropic |
| Clasificación simple | Haiku 4, GPT-4o-mini, Llama 3.3 8B | Anthropic, OpenAI, Groq |
| Velocidad + calidad razonable | Llama 3.3 70B, Qwen 3 32B | Groq, Cerebras |

**Estrategia:** Empieza con el modelo más barato que cumpla el umbral de calidad. Escala a modelos caros solo si es necesario.

**2. Latencia y throughput**

| Aplicación | Requisito | Proveedor recomendado |
|---|---|---|
| Chat interactivo | TTFT < 200 ms | **Groq** (50 ms) |
| Generación de reportes largos | Throughput > 500 tok/s | **Cerebras** (2,000 tok/s) |
| Batch processing | Latencia no crítica | OpenAI Batch API (50% descuento) |
| Agente multi-paso | Latencia acumulada baja | **Groq** o **Cerebras** |
| Voz en tiempo real | TTFT < 100 ms + streaming | **Groq** |

**Benchmark real (Llama 3.3 70B):**

| Proveedor | TTFT | Throughput | Latencia total (500 tokens) |
|---|---|---|---|
| OpenAI (vía Ollama) | 300 ms | 80 tok/s | 6.5 s |
| Groq | 50 ms | 280 tok/s | 1.8 s |
| Cerebras | 100 ms | 2,200 tok/s | 0.3 s |
| Together AI | 200 ms | 150 tok/s | 3.5 s |

**3. Costo**

**Pricing por 1M tokens (input/output):**

| Proveedor | Modelo | Input | Output | Ratio |
|---|---|---|---|---|
| OpenAI | GPT-4o | $2.50 | $10.00 | 4x |
| OpenAI | GPT-4o-mini | $0.15 | $0.60 | 4x |
| Anthropic | Claude Opus 4 | $15.00 | $75.00 | 5x |
| Anthropic | Claude Haiku 4 | $0.80 | $4.00 | 5x |
| Google | Gemini 2.5 Flash | $0.075 | $0.30 | 4x |

| **Cerebras** | Llama 3.3 70B | **$0.60** | **$0.60** | **1x** |

**Ahorro potencial:**

Migrar de GPT-4o-mini a Llama 3.3 70B en Groq:
- Input: $0.15 → $0.59 (4x más caro)
- Output: $0.60 → $0.79 (1.3x más caro)

Pero si tu tarea es 80% output (ej. generación de contenido):
- GPT-4o-mini: 0.2 × $0.15 + 0.8 × $0.60 = $0.51 / 1K tokens
- Groq Llama: 0.2 × $0.59 + 0.8 × $0.79 = $0.75 / 1K tokens

### Estrategia de cascada (cascade)

En lugar de elegir UN proveedor, usa **varios en cascada**:

```python
def generar_respuesta(prompt):
    # Intento 1: Modelo barato y rápido
    try:
        return groq_client.chat(prompt, model='llama-3.3-70b')
    except (RateLimitError, ServiceUnavailable):
        pass
    
    # Intento 2: Fallback a otro proveedor rápido
    try:
        return cerebras_client.chat(prompt, model='llama-3.3-70b')
    except:
        pass
    
    # Intento 3: Fallback a OpenAI (más caro pero confiable)
    return openai_client.chat(prompt, model='gpt-4o-mini')
```

**Ventajas:**
- Maximiza uptime (si un proveedor cae, sigues funcionando)
- Optimiza costo (usa el más barato primero)
- Reduce latencia promedio

````python
# Matriz de decisión para selección de proveedor
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Definir pesos según tipo de aplicación
aplicaciones = {
    'Chat interactivo': {'calidad': 0.25, 'latencia': 0.40, 'costo': 0.20, 'privacidad': 0.05, 'lock_in': 0.10},
    'Análisis de documentos': {'calidad': 0.40, 'latencia': 0.10, 'costo': 0.30, 'privacidad': 0.15, 'lock_in': 0.05},
    'Generación de contenido': {'calidad': 0.35, 'latencia': 0.15, 'costo': 0.35, 'privacidad': 0.05, 'lock_in': 0.10},
    'Clasificación batch': {'calidad': 0.20, 'latencia': 0.05, 'costo': 0.50, 'privacidad': 0.15, 'lock_in': 0.10},
}

# Scores por proveedor (1-5, donde 5 es mejor)
scores = pd.DataFrame({
    'OpenAI': [5, 3, 2, 2, 1],
    'Anthropic': [5, 3, 2, 2, 1],
    'Gemini': [4, 3, 4, 3, 2],
    'Groq': [4, 5, 4, 4, 5],
    'Cerebras': [4, 5, 4, 4, 5],
    'Ollama Cloud': [3, 3, 4, 5, 5],
}, index=['calidad', 'latencia', 'costo', 'privacidad', 'lock_in'])

print("="*70)
print("MATRIZ DE SCORES POR PROVEEDOR (1-5, mayor es mejor)")
print("="*70)
print(scores)
print()

# Calcular scores ponderados para cada aplicación
resultados = {}
for app_name, pesos in aplicaciones.items():
    pesos_series = pd.Series(pesos)
    scores_ponderados = scores.mul(pesos_series, axis=0).sum()
    resultados[app_name] = scores_ponderados

resultados_df = pd.DataFrame(resultados).T

print("="*70)
print("SCORES PONDERADOS POR TIPO DE APLICACIÓN")
print("="*70)
print(resultados_df.round(2))
print()

# Visualización 1: Heatmap de scores ponderados
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

for idx, (app_name, ax) in enumerate(zip(aplicaciones.keys(), axes.flatten())):
    pesos = pd.Series(aplicaciones[app_name])
    scores_ponderados_detalle = scores.mul(pesos, axis=0)
    
    sns.heatmap(scores_ponderados_detalle, annot=True, fmt='.2f', cmap='RdYlGn',
                vmin=0, vmax=2, ax=ax, cbar_kws={'label': 'Score ponderado'})
    ax.set_title(f'{app_name}', fontsize=12, fontweight='bold')
    ax.set_xlabel('Proveedor')
    ax.set_ylabel('Dimensión')

plt.suptitle('Scores ponderados por dimensión y tipo de aplicación', 
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

# Visualización 2: Ranking por aplicación
fig, ax = plt.subplots(figsize=(12, 6))

x = np.arange(len(resultados_df.columns))
width = 0.15
colores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

for i, app_name in enumerate(resultados_df.index):
    offset = width * (i - len(resultados_df.index)/2 + 0.5)
    bars = ax.bar(x + offset, resultados_df.loc[app_name], width, 
                   label=app_name, alpha=0.8)
    
    # Anotar el mejor proveedor para cada app
    best_idx = resultados_df.loc[app_name].idxmax()
    best_val = resultados_df.loc[app_name].max()
    best_pos = list(resultados_df.columns).index(best_idx)
    ax.text(best_pos + offset, best_val + 0.1, '★', 
            ha='center', fontsize=16, color=colores[i])

ax.set_ylabel('Score ponderado')
ax.set_title('Comparación de proveedores por tipo de aplicación\\n(★ = mejor opción)', 
             fontsize=13, fontweight='bold')
ax.set_xticks(x)
ax.set_xticklabels(resultados_df.columns, rotation=45, ha='right')
ax.legend(loc='upper left')
ax.grid(True, alpha=0.3, axis='y')
ax.set_ylim(0, 5)

plt.tight_layout()
plt.show()

# Recomendaciones
print("="*70)
print("RECOMENDACIONES POR TIPO DE APLICACIÓN")
print("="*70)

for app_name in resultados_df.index:
    ranking = resultados_df.loc[app_name].sort_values(ascending=False)
    print(f"\n{app_name}:")
    print(f"  1. {ranking.index[0]:<15} (score: {ranking.iloc[0]:.2f})")
    print(f"  2. {ranking.index[1]:<15} (score: {ranking.iloc[1]:.2f})")
    print(f"  3. {ranking.index[2]:<15} (score: {ranking.iloc[2]:.2f})")

print("\n" + "="*70)
print("💡 Observaciones:")
print("   • Chat interactivo: Groq/Cerebras ganan por latencia baja")
print("   • Análisis de documentos: Gemini gana por contexto largo + costo")
print("   • Generación de contenido: Groq/Cerebras por balance costo/velocidad")
print("   • Clasificación batch: Gemini/Groq por costo bajo")
print("   • OpenAI/Anthropic: Mejores en calidad pura, pero más caros y lentos")
print("="*70)
````
